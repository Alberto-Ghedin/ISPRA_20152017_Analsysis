```python
import netCDF4
import numpy as np
import seaborn as sns
import rasterio
from rasterio.crs import CRS
import pandas as pd
from datetime import date
from datetime import timedelta
```

```python
locations = pd.read_csv("/mnt/d/PhD_coding/sample_env_data_from_Copernicus/locations.csv")
coord = [ (x,y) for x,y in 
         zip(locations["Longitude"],locations["Latitude"])]
```

```python
ds = rasterio.open("./PhD_coding/med-cmcc-sal-rean-m_1674823300707.nc", "r")
ds_net = netCDF4.Dataset("./PhD_coding/med-cmcc-sal-rean-m_1674823300707.nc", "r")
```

```python
print(ds_net.variables.keys())
print(ds_net.variables["time"])
print(ds_net.variables["so"])
print(ds.tags(1))
```

```python
raster_sample = np.array(list(ds.sample(coord)))
```

```python
df_sampled_values = pd.DataFrame(columns = ["Longitude",
                                           "Latitude", 
                                           "Depth", 
                                           "Time",
                                           "Salinity"])
```

```python
depths = []
times = []
for indx in range(1,len(ds.indexes)+1): 
    depths.append(float(ds.tags(indx)["NETCDF_DIM_depth"]))
    times.append(str(
                    date(1900,1,1) + 
                        timedelta(
                            minutes=int(ds.tags(indx)["NETCDF_DIM_time"])
                                )
                    )
                )
N_bands = len(depths)
for i, sampling in enumerate(raster_sample):
    sampling[np.where(sampling == 1e20)] = np.nan
    df_sampled_values = pd.concat([
        df_sampled_values, 
        pd.DataFrame(
        {
            "Longitude" : [coord[i][0]] * N_bands,
            "Latitude" : [coord[i][1]] * N_bands, 
            "Depth" : depths, 
            "Time" : times,
            "Salinity" : sampling
        })
    ])
```

```python
df_sampled_values.dropna(subset=["Salinity"]).to_csv("sampled_salinity.csv", index=False)
```

```python
ds_net.close()
ds.close()
```
